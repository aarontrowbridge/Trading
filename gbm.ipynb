```python
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
tsla = yf.Ticker('TSLA')
```

```python
hist = tsla.history(period="5y")
```

```python
hist.info
```

```python
plt.plot(hist['Close'])
```

```python
highs = hist['High']
lows = hist['Low']
closes = hist['Close']
vols = (highs - lows) / closes
```

```python
plt.plot(vols)
```

```python
def my_histogram(vals, nbars):
    maxval = max(vals)
    minval = min(vals)
    width = (maxval - minval) / nbars
    histogram = np.zeros(nbars)
    for val in vals:
        bar = (val - minval) // width 
        histogram[int(bar) - 1] +=1
    return np.arange(minval, maxval, width), histogram
```

```python
vol_axis, histo = my_histogram(vols, 200)
len(vol_axis)
```
